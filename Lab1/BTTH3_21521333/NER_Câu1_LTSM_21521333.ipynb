# english data

```python
import pandas as pd
data=pd.read_csv('/content/DS310/Lab1/data/ner_dataset.csv',encoding='latin1')
data = data.fillna(method="ffill")

import pandas as pd
from sklearn.model_selection import train_test_split
import os
os.makedirs('data',exist_ok=True)
unique_sentences = data['Sentence #'].unique()

# Chia tập index của các câu thành tập train, dev và test
train_sent, temp_sent = train_test_split(unique_sentences, test_size=0.3, random_state=42)
dev_sent, test_sent = train_test_split(temp_sent, test_size=0.5, random_state=42)

# Lọc các dòng của DataFrame dựa trên tập index đã chia
train_data = data[data['Sentence #'].isin(train_sent)]
dev_data = data[data['Sentence #'].isin(dev_sent)]
test_data = data[data['Sentence #'].isin(test_sent)]

# In thông tin về số lượng mẫu trong mỗi tập
print(f"Train set size: {len(train_data)} samples")
print(f"Dev set size: {len(dev_data)} samples")
print(f"Test set size: {len(test_data)} samples")

print(len(train_data['Sentence #'].unique()))
print(len(dev_data['Sentence #'].unique()))
print(len(test_data['Sentence #'].unique()))

train_data.to_csv('data/train.csv', index=False)
dev_data.to_csv('data/dev.csv', index=False)
test_data.to_csv('data/test.csv', index=False)
```
