# Pho NER

```python
import os
import pandas as pd

os.makedirs('data',exist_ok=True)

train=pd.read_csv('/content/DS310/Lab1/data/train_phoNER_COVID10.csv',encoding='utf-8')
train=train.rename(columns={'sentence #':'Sentence #','tag':'Tag','word':'Word'})
train['POS']=train['Tag']
train.to_csv('data/train.csv',index=False)

dev=pd.read_csv('/content/DS310/Lab1/data/dev_phoNER_COVID19.csv',encoding='utf-8')
dev=dev.rename(columns={'sentence #':'Sentence #','tag':'Tag','word':'Word'})
dev['POS']=dev['Tag']
dev.to_csv('data/dev.csv',index=False)

test=pd.read_csv('/content/DS310/Lab1/data/test_phoNER_COVID19.csv',encoding='utf-8')
test=test.rename(columns={'sentence #':'Sentence #','tag':'Tag','word':'Word'})
test['POS']=test['Tag']
test.to_csv('data/test.csv',index=False)

print(train.shape)
print(dev.shape)
print(test.shape)
```
